```

df = df.drop_duplicates()
```

```

df.columns

# Before renaming:
# 'math score', 'reading score', 'writing score'
```

```

df.columns = [
    'gender',
    'race_ethnicity',
    'parent_education',
    'lunch',
    'test_prep',
    'math_score',
    'reading_score',
    'writing_score'
]

# Example:
# Aarav → math_score = 72
```

```

df[(df['math_score'] < 0) | (df['math_score'] > 100)]

# Example:
# Aarav math = 72 ✔
# Rohan math = 90 ✔
```

```

df[(df['reading_score'] < 0) | (df['writing_score'] > 100)]
```

```

df['gender'].value_counts()

# Example:
# Aarav → male
# Ananya → female
```

```

df['race_ethnicity'].unique()

# Example:
# group B, group C, group E
```

```

df['test_prep'].value_counts()

# Example:
# Ananya → completed
# Aarav → none
```

```

df[['math_score','reading_score','writing_score']].describe()
```

```

df['average_score'] = (
    df['math_score'] +
    df['reading_score'] +
    df['writing_score']
) / 3

# Example:
# Aarav  → 72.67
# Ananya→ 82.33
# Priya → 55.67
```

```

df[['math_score','reading_score','writing_score','average_score']].head()
```

```

def performance(avg):
    if avg >= 75:
        return 'High'
    elif avg >= 50:
        return 'Medium'
    else:
        return 'Low'

df['performance_level'] = df['average_score'].apply(performance)

# Example:
# Rohan → High
# Aarav → Medium
# Low scorer → Low
```

```

df[['average_score','performance_level']].head()
```

```

plt.figure(figsize=(8,5))
sns.boxplot(data=df[['math_score','reading_score','writing_score']])
plt.show()
```
