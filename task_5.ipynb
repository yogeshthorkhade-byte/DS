```python
#Write a NumPy program to find the sum of all elements in an array [10, 20, 30]
arr1 = np.array([10, 20, 30])
print(np.sum(arr1))
```

```python
#Write a NumPy program to find the maximum element in [5, 12, 3, 9].
arr3 = np.array([5, 12, 3, 9])
print(np.max(arr3))
```

```python
#Write a NumPy program to add two arrays [1, 2, 3] and [4, 5, 6].
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print(a + b)
```
